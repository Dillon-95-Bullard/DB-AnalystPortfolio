```python
import pandas as pd
tipsAnalysis = pd.read_csv('tipsData.csv')    # Creates a dataframe stored as tipsAnalysis
tipsAnalysis.describe(include='all')     # Equilavent to using the DataAnalysis tool in Excel
```

```python
tipsAnalysis['tipRate'] = tipsAnalysis['tip'] / tipsAnalysis['total_bill']
print(tipsAnalysis.shape)                                                             
tipsAnalysis.head(10)
```

```python
tipsAnalysis.info()
```

```python
newTipsAnalysis = pd.get_dummies(tipsAnalysis, prefix=None, prefix_sep='_')
newTipsAnalysis.describe(include='all')
```

```python
newTipsAnalysis['tipRate'].median()
```

```python
fltMedian = newTipsAnalysis['tipRate'].median()
```

```python
print('The observed median tipRate is ' + str(fltMedian) + '%.')
```

```python
xBar = newTipsAnalysis['total_bill'].mean()
print(xBar)
s = newTipsAnalysis['total_bill'].std()
print(s)
n = newTipsAnalysis['total_bill'].count()
print(n)
Z = 1.96     # Z-score corresponding to a 95% level confidence interval
fltStdErr = s/(n**.5)
print(fltStdErr)
# Now, let's determine the 95% confidence estimate for average total_bill
fltCI_Lower = xBar - Z*fltStdErr
print(fltCI_Lower)
fltCI_Upper = xBar + Z*fltStdErr
print(fltCI_Upper)
```
